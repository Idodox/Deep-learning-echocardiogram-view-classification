```python
def clean_list(folder_list):
    clean_list = []
    for file in folder_list:
        if file == '.DS_Store' or file[-4:] == '.dcm':
            continue
        clean_list.append(file)
    return clean_list

def add_dcm_extension_to_files_in_folder(folder):
#     print("adding dcm extension to files in folder", folder)
    folder = Path(folder)
    for file in folder.iterdir():
        # Validate contents are files not folders
        if file.is_file():
            current_path = str(file)
            new_path = str(file) + '.dcm'
            os.rename(current_path, new_path)
        else: 
            print("Something is weird, tried to add dcm to folder")
            print(folder)
```

```python
data_dir = '/Users/idofarhi/Documents/Thesis/to_label/temp'

patients = clean_list(os.listdir(data_dir))

for patient in patients:
    ### extract studies from date folder, if it(they) are in one
    # now we expect to see in folder_list a folder with day of month
#     print('Processing folder', patient)
    folder_list = clean_list(os.listdir(os.path.join(data_dir, patient)))
    for DOM_folder in folder_list:
        try:
            if int(DOM_folder) < 32:
                study_list = clean_list(os.listdir(os.path.join(data_dir, patient, DOM_folder)))
                for study in study_list:
                    current_path = os.path.join(data_dir, patient, DOM_folder, study)
                    new_path = os.path.join(data_dir, study)
                    add_dcm_extension_to_files_in_folder(current_path)
                    shutil.move(current_path, new_path)
                    
        except Exception:
            print("Exception, likely no date folder found for patient", patient)
            print("Day of month folder:", DOM_folder)
            
            if os.path.isdir(os.path.join(data_dir, patient, DOM_folder)):
                print(DOM_folder, "is a folder. Trying to move it under patient")
                current_path = os.path.join(data_dir, patient, DOM_folder)
                new_path = os.path.join(data_dir, DOM_folder)
                add_dcm_extension_to_files_in_folder(current_path)
                shutil.move(current_path, new_path)
#             traceback.print_exc()
```
