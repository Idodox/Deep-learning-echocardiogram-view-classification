```python
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from skimage.io import imsave
from cv2 import cvtColor, COLOR_BGR2GRAY
```

```python
def load_file(filename):
    dicom = pydicom.read_file(filename)
    vid = dicom.pixel_array
    return vid

def write_frames(img_array, video_name, frames_dir, convert_to_grayscale = True):
    for i, frame in enumerate(img_array):
        output_filename = frames_dir + '/' + video_name[:-4] + "_" + str(i) + '.jpg'
#             output_filename = output_filename.encode('unicode_escape')
        if convert_to_grayscale:
            frame = cvtColor(frame, COLOR_BGR2GRAY)
        print(frame.shape)
        break
        imsave(output_filename,frame)
```

```python
classes = {'apex', 'mitral', 'papillary'}
for view in classes:
    parent_dir = os.path.dirname(os.getcwd())
    frames_dir = os.path.join(parent_dir, "Data/frames", view)
    dicom_dir = os.path.join(parent_dir, "Data", view, "dicom")
    video_list = os.listdir(dicom_dir)
    if '.DS_Store' in video_list: video_list.remove('.DS_Store')
        
    # Validate frames_dir exists:
    if not os.path.exists(frames_dir):
        print("directory doesn't exist, creating it...")
        print(frames_dir)
        os.makedirs(frames_dir)
    
    # Go through each video and save all frames
    for video_name in tqdm(video_list):
        video = load_file(os.path.join(dicom_dir, video_name))
        write_frames(video, video_name, frames_dir)
```
